```
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
titanic_data = sns.load_dataset("titanic")

# Dropping rows with missing values for simplicity
titanic_data = titanic_data.dropna()

print(titanic_data.head())
```

```
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic_data)
plt.title('Age vs Fare Scatter Plot')
plt.show()
```

```
average_fare_by_age = titanic_data.groupby('age')['fare'].mean()

plt.figure(figsize=(10, 6))
average_fare_by_age.plot()
plt.title('Average Fare by Age Line Graph')
plt.xlabel('Age')
plt.ylabel('Average Fare')
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['age'], bins=20, kde=False)
plt.title('Age Histogram')
plt.xlabel('Age')
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.boxplot(x='survived', y='fare', data=titanic_data)
plt.title('Fare Box and Whisker Plot')
plt.show()
```

```
correlation_matrix = titanic_data[['age', 'fare', 'sibsp', 'parch']].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
